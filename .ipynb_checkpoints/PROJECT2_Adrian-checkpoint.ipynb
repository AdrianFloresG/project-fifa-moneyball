```python
fifa_country['Value1'] = fifa_country['Value'].str.replace("€","")
fifa_country['Value1']
fifa_country['Wage1'] = fifa_country['Wage'].str.replace("€","")
fifa_country['Wage1']
```

```python
def extract(col):
    
    x=col[:-1]
    y=col[-1]
    if y=='K':
        return float(x)*1000
    elif y=='M':
        return float(x)*1000000
```

```python
fifa_country['Value'] = fifa_country['Value1'].apply(extract).fillna(0)
fifa_country['Wage'] =fifa_country['Wage1'].apply(extract).fillna(0)
fifa_country
```

```python
def log_transform(num):
    return np.log(num)/np.log(2)  
```

```python
fifa_country['Value_t'] = fifa_country['Value'].apply(log_transform)
fifa_country['Value_t'].hist()
```

```python
from sklearn.preprocessing import StandardScaler

standardizer = StandardScaler()
fifa_country['Value_s'] = standardizer.fit_transform(fifa_country[['Value_t']])
fifa_country['Value_s']
```
